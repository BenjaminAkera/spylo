# GeoCoding the collected data

```python
import simplekml
import os
```

```python
data=pd.read_csv("output.csv")
file = simplekml.Kml()

for Long,Lat,name in zip(data["Lat"],data["Long"],data["name"]):
    file.newpoint(coords=[(Lat,Long)],name=(name))
file.save(os.path.expanduser("file.kml"))
```

# Trying Folium

```python
import os
import folium

spots = pd.read_csv("output.csv", usecols=["Lat","Long","name"])
locations = spots[['Lat', 'Long']]
locationlist = locations.values.tolist()
len(locationlist)
```

```python
import pandas as pd


state_geo = os.path.join('data', 'states.json')
state_unemployment = os.path.join('data', 'unemployment.csv')

state_data = pd.read_csv(state_unemployment)


m = folium.Map(
    location=[0.3292,32.5710],
    zoom_start=16,
    tiles='OpenStreetMap'
)

for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point],popup=spots['name'][point]).add_to(m)
    #folium.CircleMarker(locationlist[point],radius=60,fill=True,popup=popup,weight=2,).add_to(m)
    
    
m.choropleth(
    state_geo,
    data=state_data,
    columns=['State', 'Unemployment'],
    key_on='feature.id',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Unemployment Rate (%)'
)


popup = 'Possible location'

folium.CircleMarker(
    location=[0.3819,32.5762],
    radius=30,
    fill=True,
    popup=popup,
    weight=2,
).add_to(m)

folium.CircleMarker(
    location=[0.3107, 32.6348],
    radius=30,
    fill=True,
    popup=popup,
    weight=2,
).add_to(m)

folium.CircleMarker(
    location=[0.2870, 32.6037],
    radius=30,
    fill=True,
    popup=popup,
    weight=2,
).add_to(m)

folium.CircleMarker(
    location=[0.3080, 32.5486],
    radius=30,
    fill=True,
    popup=popup,
    weight=2,
).add_to(m)

fg= folium.FeatureGroup(name='My Map')

fg.add_child(folium.GeoJson(open("KAMPALA.geojson",encoding = "utf-8-sig").read()))

m.add_child(fg)

m.save(os.path.join('results', 'CheckZorder.html'))

m
```
